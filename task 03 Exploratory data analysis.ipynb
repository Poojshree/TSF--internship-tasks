## Data visualization

```python
sns.pairplot(store)
```

## plot of products in each category

```python
plt.figure(figsize=(10,8))
store['Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

## plot of products in each Sub-category

```python
plt.figure(figsize=(10,8))
store['Sub-Category'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

## Total profit and sales per sub category

```python
plt.figure(figsize=(8,6))
store.groupby('Sub-Category')['Profit','Sales'].agg(['sum']).plot.bar(figsize=(16,10))
plt.title('Total Profit and Sales per Sub-Category')
plt.show();
```

## Count of Sub-Category region wise

```python
plt.figure(figsize=(20,10))
sns.countplot(x="Sub-Category", hue="Region", data=store)
plt.show()
```

## Checking the Profit based on the Region :

```python
store.groupby('Region')['Profit'].agg(['sum']).sort_values(['sum'], ascending=False).plot.bar()
```

## profit based on each state

```python
store.groupby('State')['Profit'].agg(['sum']).sort_values(['State'], ascending=False).plot.bar(figsize=(16,10))
```

## most profit in subcategory

```python
store.groupby('Sub-Category')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(16,10));
```

## profit in shipping mode 

```python
store.groupby('Ship Mode')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(16,10));
```

## most profit in each segment

```python
store.groupby('Segment')['Profit'].agg(['sum']).sort_values(['sum'], 
                                                         ascending=False).plot.bar(figsize=(16,10));
```
